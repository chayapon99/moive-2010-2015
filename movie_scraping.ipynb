```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.common.action_chains import ActionChains
from itertools import chain
import time

film_url = 'https://www.scrapethissite.com/pages/ajax-javascript/#'
year_list = ['2010', '2011', '2012', '2013', '2014', '2015']
counter = 0
url_list = [film_url + year for year in year_list]

film_list = []

for url in url_list:
    # Set up the Chrome webdriver
    chrome_service = ChromeService()  # Set the path to your chromedriver executable
    driver = webdriver.Chrome(service=chrome_service)
    
    # Load the website
    driver.get(url)
    
    # You may need to wait for dynamic content to load
    clickable = driver.find_element(By.ID, year_list[counter])
    ActionChains(driver).click(clickable).perform()
    
    # Wait for some amount of time for DOM to be loaded
    time.sleep(10)
    
    # Use Selenium to find and print the content
    try:    
        # Example: Extracting text from an element with class 'example-class'
        thead_tag = driver.find_elements(By.TAG_NAME, 'thead')  
        thead_list = list(chain.from_iterable(value.text.split() for value in thead_tag))
        
        # # Find all tr elements with the class 'film' within the tbody
        tbody_element = driver.find_element(By.ID, 'table-body')
        film_rows = tbody_element.find_elements(By.XPATH, "//tr[@class='film']")
        for film_row in film_rows:
            film_list.append(film_row.find_element(By.CLASS_NAME, 'film-title').text)
            film_list.append(year_list[counter])
            film_list.append(film_row.find_element(By.CLASS_NAME, 'film-nominations').text)
            film_list.append(film_row.find_element(By.CLASS_NAME, 'film-awards').text)
            
    except Exception as e:
        print(f"Error: {e}")
    finally:
        # Close the browser
        driver.close()
        counter += 1

print(film_list)
```

```python
resized_film_list = [film_list[i:i+4] for i in range(0, len(film_list), 4)]
resized_film_list
```

```python
import pandas as pd

# Create a DataFrame
df = pd.DataFrame(resized_film_list)

# Rename the columns
df.columns = ['title', 'year','nominations', 'awards']

# Display the DataFrame
print(df)

# Save the data
df.to_csv('movie_by_year.csv')
```
